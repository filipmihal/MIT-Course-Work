```python
#Binary search
def search(A, x):
    u = len(A)-1
    l = 0
    while u >= l:
        searchIndex = int((u+l)/2) 
        print(searchIndex)
        if A[searchIndex] == x:
            return searchIndex
        elif A[searchIndex] > x:
            u = searchIndex - 1
        else:
            l = searchIndex + 1
```

```python
myList = [*range(0,100)]
searchedNumber = -1
print(search(myList, searchedNumber))
```
